```python
oil_price = yf.Ticker("CL=F")
oil_df = oil_price.history(period="10y")

# Convert to DataFrame and plot
oil_df = pd.DataFrame(oil_df)
plt.figure(figsize=(12, 6))
oil_df.plot(y='Close', use_index=True, title='WTI Crude Oil Price History')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)

# Display first few rows
print("Oil price data head:")
oil_df.head()
```

```python
# Get USD Index data
usd_price = yf.Ticker("DX-Y.NYB")
usd_df = usd_price.history(period="10y")

# Convert to DataFrame and plot
usd_df = pd.DataFrame(usd_df)
plt.figure(figsize=(12, 6))
usd_df.plot(y='Close', use_index=True, title='US Dollar Index History')
plt.xlabel('Date')
plt.ylabel('USD Index Value')
plt.grid(True)

# Display first few rows
print("USD Index data head:")
usd_df.head()
```

```python
df = gold_price.history(period="1y")
oil_df = oil_price.history(period="1y")
usd_df = usd_price.history(period="1y")

# Normalized the closing prices to start at 100
gold_normalized = df['Close'] / df['Close'].iloc[0] * 100
oil_normalized = oil_df['Close'] / oil_df['Close'].iloc[0] * 100
usd_normalized = usd_df['Close'] / usd_df['Close'].iloc[0] * 100

plt.figure(figsize=(15, 8))
plt.plot(gold_normalized.index, gold_normalized, label='Gold', color='gold')
plt.plot(oil_normalized.index, oil_normalized, label='Oil', color='brown')
plt.plot(usd_normalized.index, usd_normalized, label='USD Index', color='green')

plt.title('Normalized Prices: Gold, Oil, and USD Index (Base=100)')
plt.xlabel('Date')
plt.ylabel('Normalized Price')
plt.grid(True)
plt.legend()      
plt.show()

#correlation matrix
correlation_df = pd.DataFrame({
    'Gold': gold_normalized,
    'Oil': oil_normalized,
    'USD': usd_normalized
})
print("\nCorrelation Matrix:")
print(correlation_df.corr())
```

```python
# Normalize the closing prices to start at 100 (not 10)
df = gold_price.history(period="5y")
oil_df = oil_price.history(period="5y")
usd_df = usd_price.history(period="5y")

gold_normalized = df['Close'] / df['Close'].iloc[0] * 100
oil_normalized = oil_df['Close'] / oil_df['Close'].iloc[0] * 100
usd_normalized = usd_df['Close'] / usd_df['Close'].iloc[0] * 100

# Create the combined plot
plt.figure(figsize=(15, 8))
plt.plot(gold_normalized.index, gold_normalized, label='Gold', color='gold')
plt.plot(oil_normalized.index, oil_normalized, label='Oil', color='brown')
plt.plot(usd_normalized.index, usd_normalized, label='USD Index', color='green')

plt.title('10-Year Normalized Prices: Gold, Oil, and USD Index (Base=100)')
plt.xlabel('Date')
plt.ylabel('Normalized Price (Base=100)')
plt.grid(True)
plt.legend()      

# Add text with date range
start_date = gold_normalized.index[0].strftime('%Y-%m-%d')
end_date = gold_normalized.index[-1].strftime('%Y-%m-%d')
plt.text(0.02, 0.98, f'Period: {start_date} to {end_date}', 
         transform=plt.gca().transAxes, 
         fontsize=9)

plt.show()


print(f"\nCorrelation Matrix ({start_date} to {end_date}):")
print(correlation_df.corr())
```

```python
#gold price last 10 days
df = gold_price.history(period="50y")
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Close'], marker='o', linestyle='-', color='gold')
plt.title('Gold Price - Last 10 Days')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```

```python
#last 10 days gold moving average
df['MA10'] = df['Close'].rolling(window=10).mean()
#plotting moving average
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Close'], label='Close Price', color='gold')
plt.plot(df.index, df['MA10'], label='10-Day MA', color='blue', linestyle='--')
plt.title('Gold Price with 10-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import pmdarima as pm
```

```python
df.index
```

```python
plot_acf(df['Close'], lags=30)
plt.title('ACF (Autocorrelation Function)')
plt.show()

plot_pacf(df['Close'], lags=30, method='ols')
plt.title('PACF (Partial Autocorrelation Function)')
plt.show()
```

```python
model = ARIMA(df['Close'], order=(2, 1, 0))  # replace with your chosen order, e.g., (1,1,1)
model_fit = model.fit()

# Forecast next 5 days
forecast = model_fit.forecast(steps=5)

# Create future dates based on current index
import pandas as pd
future_dates = pd.date_range(df.index[-1] + pd.Timedelta(days=1), periods=5)

# Prepare result DataFrame
result_df = pd.DataFrame({'Date': future_dates, 'Forecasted_Close': forecast.values})

print(result_df)
```

```python
last_close = df['Close'].iloc[-1]

# Calculate percentage increase for each predicted day
result_df['Percent_Increase'] = ((result_df['Forecasted_Close'] - last_close) / last_close) * 100


result_df['Percent_Increase'] = result_df['Percent_Increase'].round(2)

print(result_df[['Date', 'Forecasted_Close', 'Percent_Increase']])
```

```python
last_date = df.index[-1]

# 1 week ago
date_1w_ago = last_date - pd.Timedelta(weeks=1)
price_1w_ago = df.loc[df.index <= date_1w_ago, 'Close'].iloc[-1]

# 1 month ago (approx 30 days)
date_1m_ago = last_date - pd.Timedelta(days=30)
price_1m_ago = df.loc[df.index <= date_1m_ago, 'Close'].iloc[-1]

# Current price
current_price = df['Close'].iloc[-1]

# Calculate percentage change
change_1w = ((current_price - price_1w_ago) / price_1w_ago) * 100
change_1m = ((current_price - price_1m_ago) / price_1m_ago) * 100

print(f"Price change in last 1 week: {change_1w:.2f}%")
print(f"Price change in last 1 month: {change_1m:.2f}%")
```
